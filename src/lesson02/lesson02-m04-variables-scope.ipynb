```python
prices = [ 100.0, 50.0, 80.0 ]
quantities = [ 3, 4, 7]

print(prices)
print(quantities)

def apply_tax(subtotal, tax_rate=0.18):
    return subtotal * (1 + tax_rate)

def final_invoice(subtotal:float, tax:float, discount=0.0) -> float:
    invoice = subtotal + tax - discount
    return invoice
    
subtotal = sum(p * q for p, q in zip(prices, quantities))
tax = apply_tax(subtotal)
total = final_invoice(subtotal, tax, 20.0)

print(f"Subtotal: {subtotal:.2f}")
print(f"Total with tax: {total:.2f}")
```
